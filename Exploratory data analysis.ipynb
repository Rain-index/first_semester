***Population and Sample***

Sampling and the general population are key concepts in statistics.

* A general population is a complete group of objects or elements that we want to get information about. For example, if we study the level of education among students at a university, then the general population will be all students of this university.
* A sample is a subset of the general population that is selected for the study. The sample should be representative so that the results of the study can be generalized to the entire general population. For example, if we interview 100 students from a university, it will be a sample.
Using sampling saves time and resources, and simplifies the data collection process, especially when the general population is very large.

# Exploratory data analysis

Exploratory data analysis (DEA) is an approach to data analysis that focuses on identifying patterns and features in data using visual and quantitative methods. The main objectives of the EDA:

* Understand the data structure and the relationships between variables.
* Identify anomalies, outliers, and missing values.
* Formulate hypotheses for further analysis.
* Prepare the data for subsequent modeling.

The main methods of EDA include:

* Data visualization (graphs, charts, histograms)
* Calculation of descriptive statistics (mean, median, standard deviation)
* Analysis of correlations between variables
* Grouping and segmentation of data
* Detection of emissions and anomalies

Exploratory data analysis helps to better understand the structure and features of data, which in turn allows for more effective application of machine learning and statistical modeling methods.

```python
# The relationship between the value of insulin levels and sugar levels, Outcome target variable (presence of diabetes)
sns.set_style('darkgrid')
plt.title('Зависимость между значением Insulin и Glucose',size=16)
plt.xlabel('Insulin',size=12)
plt.ylabel('Glucose',size=12)
sns.scatterplot(x='Insulin', y='Glucose', data=data, hue='Outcome', edgecolor='black', palette='cubehelix', hue_order=[0,1])
plt.show()
```
